```
import re
import pandas as pd


def build_decks(text):
  decks = {}
  category = ''
  for l, line in enumerate(text.splitlines()):
    if len(line.split()) == 1:
      name = line
      deck = {}
    elif line.endswith(')'):
      if category:
        deck.update({category: {'count': catcount, 'cards': cards}})
      category = line.split()[0]
      catcount = int(re.findall('(\d+)', line)[0])
      cards = {}
    elif line.startswith((' ', 'DECKLIST', 'SORT')):
      pass
    elif line and line[0].isnumeric():
      if category == 'Sideboard':
        sb = []
        card = []
        for i in line.split():
          if i.isnumeric():
            if card:
              sb.append(card)
            card = [i]
          else:
            card.append(i)
        sb.append(card)
        for card in sb:
          num = int(card.pop(0))
          cards.update({' '.join(card): num})
      else:
        parts = line.split()
        count = int(parts.pop(0))
        if count < 45:
          cards.update({' '.join(parts): count})
    elif not line:
      try:
        deck.update({category: {'count': catcount, 'cards': cards}})
      except:
        print(l, line)
      decks.update({name: deck})
  for deck in decks.values():
    for cat in deck.values():
      assert sum(cat['cards'].values()) == cat['count']
  return decks
```

```
decks = build_decks(text)
```

```
cardlist = {}
sideboards = {}
for deck in decks.values():
  for name, cat in deck.items():
    if name != 'Sideboard':
      for card, count in cat['cards'].items():
        if card not in cardlist.keys():
          cardlist.update({card: {'category': name, 'counts': count, 'decks': 1}})
        else:
          cardlist[card]['counts'] += count
          cardlist[card]['decks'] += 1
    else:
      for card, count in cat['cards'].items():
        if card not in sideboards.keys():
          sideboards.update({card: {'category': name, 'counts': count, 'decks': 1}})
        else:
          sideboards[card]['counts'] += count
          sideboards[card]['decks'] += 1
```

```
totals = pd.DataFrame(cardlist).T
totals['mean/deck'] = totals.counts.divide(totals.decks)
totals.sort_values(by=['mean/deck', 'counts'], ascending=False)
```
